```python
import numpy as np
import numpy
import tensorflow as tf
from keras.models import Sequential
from keras.layers import ReLU
from keras.layers import Dense
from matplotlib import pyplot
from pandas import read_csv
from scipy import stats
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
class ts_preprocessor(object):
    
    def __init__(self):
        self.scaleFactor = 1
    
    def fit_transform(self, data):
        self.scaleFactor = data.max() + 1 if self.scaleFactor == 1.0 else self.scaleFactor
        return np.array([(x / self.scaleFactor) for x in data])
    
    def inverse_transform(self, data):
        return [(x * self.scaleFactor) for x in data]
```

## Neural Network model

```python
data = read_csv('rht_new.csv', header = 0)
data
```

```python
values = data.values
```

```python
# make numeric
values = values.astype('float32')
```

```python
#make scaler invariant (remove scale factor)
#fit transform over tranform, so I can develop a scaler model from the data
#to be used to reverse scale invariance
scaler = ts_preprocessor()
values = scaler.fit_transform(values)
```

```python
# set examples and target
examples = values[:,0:2]
target = values[:,2:3]
```

```python
c = len(examples[0])
test_s_index = int(len(values) * 0.8)
```

```python
# build the neural network model
model = Sequential()
model.add(Dense(10, input_shape=(c,), activation= tf.nn.sigmoid))
model.add(Dense(20, input_shape=(c,), activation= tf.nn.sigmoid))
outputlayer = Dense(units = 1, activation= tf.nn.sigmoid)
model.add(outputlayer)
```

```python
model.compile(loss=tf.losses.mean_squared_error, optimizer='adam')
```

```python
validation_ex = examples[test_s_index:,]
validation_tgt = target[test_s_index:,]
```

```python
# train
ret = model.fit(examples, target, batch_size= 50, epochs = 2000)
```

```python
# evaluate model performance
vret = model.evaluate(validation_ex, validation_tgt)
print("val test loss, val test acc: ", vret)
```

## Polynomial Regression model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('rht_new.csv')
X = df[['side 1','side 2']]
y = df['side 3']
```

```python
model_linear = LinearRegression()
model_linear.fit(X, y)
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(X)
model_linear_2 = LinearRegression()
# fit the model
model_linear_2.fit(X_poly, y)
y_poly_pred = model_linear_2.predict(X_poly)
```

## Paired t-test to compare the performance of 2 prediction model

### Test 1: The test dataset is not significantly different from the training data

```python
# Create a dataset with side 1 and side 2
new_data = {
    'side 1': np.random.randint(5, 1000, 500),  # Generate random values for side 1
    'side 2': np.random.randint(5, 1000, 500)   # Generate random values for side 2
}

# Calculate side 3 using the Pythagorean theorem
new_data['side 3'] = np.sqrt(new_data['side 1']**2 + new_data['side 2']**2)

# Create a DataFrame
test_df = pd.DataFrame(new_data)
test_df
```

```python
# Create empty arrays to store RMSE values
rmse_nn_array = []
rmse_poly_array = []

for i in range(len(test_df)):
    # Get the side1, side2, and side3 values for the current row
    side1 = test_df.loc[i, 'side 1']
    side2 = test_df.loc[i, 'side 2']
    side3 = test_df.loc[i, 'side 3']
    
    # Make predictions for the current row with the Neural Network model
    scaled_input_nn = scaler.fit_transform(np.array([[side1, side2]]))
    prediction_nn = model.predict(scaled_input_nn)
    prediction_nn = scaler.inverse_transform(np.array([[prediction_nn[0][0], 0]]))[0][0]
    
    # Make predictions for the current row with the Polynomial Regression model
    feature_poly = poly_reg.transform(np.array([[side1, side2]]))
    prediction_poly = model_linear_2.predict(feature_poly)[0]
    
    # Calculate RMSE for the current row and store it in the arrays
    rmse_nn = np.sqrt(metrics.mean_squared_error([side3], [prediction_nn]))
    rmse_poly = np.sqrt(metrics.mean_squared_error([side3], [prediction_poly]))
    
    rmse_nn_array.append(rmse_nn)
    rmse_poly_array.append(rmse_poly)

print('Done')
```

```python
# Perform t-test
t_stat, p_value = stats.ttest_rel(rmse_nn_array, rmse_poly_array)
print(t_stat, p_value)
```

```python
#the significance level (alpha)
alpha = 0.05 

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the models.")
    if np.mean(rmse_nn) < np.mean(rmse_poly):
        print("The Neural Network model is better (lower RMSE).")
    else:
        print("The Polynomial Regression model is better (lower RMSE).")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the models.")
```
